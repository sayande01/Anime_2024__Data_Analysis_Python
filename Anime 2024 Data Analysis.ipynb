```python
#Importing dependencies
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import dataset
df = pd.read_csv(r'anime.csv')
```

### Exploratory Data Analysis

```python
#top 5 rows
df.head()
```

```python
#row-col
df.shape
```

```python
#Details
df.info()
```

```python
#Check Duplicate
duplicate_rows = df.duplicated()
print(df[duplicate_rows])
```

```python
df.columns
```

```python
# Year Column
df['year']=df['Premiered'].str[-4:]
df.head()
```

```python
# Top 10 Anime

df1= df[['Popularity','English']].sort_values(by = 'Popularity', ascending = True).head(10)
df1
```

```python
sns.set()
```

```python
plt.figure(figsize = (10,5))
plt.xticks(rotation=45)
sns.barplot(x = df1['English'], y = df1['Popularity'], palette = 'dark' )
plt.show()
```

```python
# Year wise releases
```

```python
df2 = df['year'].value_counts().sort_values(ascending= False).head(10)
df2
```

```python
sns.barplot(x = df2.values, y = df2.index)
plt.xlabel('No of Releases')
plt.title('Year wise releases')
plt.show()
```

```python
# Score distribution
```

```python
plt.figure(figsize = (8,5))
sns.histplot(x = df['Score'],kde = True, bins = 25)
plt.xlabel('Score')
plt.title('Score distribution')
plt.show()
```

```python
# Yearwise Members
```

```python
df4 = df.groupby('year')['Members'].sum().sort_values(ascending = False).head(7)
df4
```

```python
df4.index.astype(int)
```

```python
df5 = df4.sort_values()
df5
```

```python
df5.plot(kind= 'line')
plt.ylabel('Total Members')
plt.title('Yearwise total member count')
plt.show()
```

```python
#Type
```

```python
df6 = df['Type'].value_counts().head(3)
df6
```

```python
df6.plot(kind = 'pie', autopct = '%1.2f%%')
plt.ylabel(None)
plt.title('Type wise distribution')
plt.show()
```

```python
#Genres
```

```python
df7 = df['Genres'].value_counts().head(5)
df7
```

```python
plt.xticks(rotation=90)
sns.barplot(x = df7.index, y = df7.values, palette = 'pastel' )
plt.ylabel('No of releases')
plt.xlabel(None)
plt.title('Genre wise No of titles')
plt.show()
```

```python
# Top Studios
```

```python
df8 = df['Studios'].value_counts().sort_values(ascending = False).head(5)
df8
```

```python
sns.barplot(x = df8.values, y = df8.index)
plt.xlabel('No of Releases')
plt.title('Studio wise releases')
plt.show()
```

```python
#Source
```

```python
df9 = df['Source'].value_counts().head(4)
df9
```

```python
df9.plot(kind = 'pie', autopct = '%1.2f%%')
plt.ylabel(None)
plt.title('Source wise distribution')
plt.show()
```

```python
# Score vs Popularity Scatter Plot
```

```python
plt.figure()
sns.scatterplot(x='Popularity', y='Score', data=df)
plt.xlabel('Popularity')
plt.ylabel('Score')
plt.title('Popularity vs Score')
plt.show()
```

```python
# Score distribution for different types of Anime
```

```python
plt.figure()
sns.boxplot(x='Type', y='Score', data=df)
plt.xlabel('Type')
plt.ylabel('Score')
plt.title('Distribution of Scores by Type')
```
